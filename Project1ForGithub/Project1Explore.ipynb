### 2)  Which degree programs are students enrolled in? 
###  Displaying all the reported degree programs.

```python
#displaying the data under 'which courses are you enrolled in' to question 2
df['What degree program are you currently enrolled in?'].value_counts()
```

### Condensing like degree programs to reduce the number of reported programs for question 2.

```python
#data cleaning by combining alike degrees
t=0  #setting index counter value for display
for degree in df['What degree program are you currently enrolled in?']:  #iterating through all of the values in the column
    t += 1  # counter variable for each iteration
    print(t)  #displaying the index number
    print(degree)  # displayint the content of the index value
        
    if 'Engineer' in degree:
        df['What degree program are you currently enrolled in?'].replace(degree, 'Engineering', inplace=True)
    elif 'engineer' in degree:
        df['What degree program are you currently enrolled in?'].replace(degree, 'Engineering', inplace=True)
    elif 'ENGINEER' in degree:
        df['What degree program are you currently enrolled in?'].replace(degree, 'Engineering', inplace=True)   
    elif 'EDAM' in degree:
        df['What degree program are you currently enrolled in?'].replace(degree, 'Engineering', inplace=True)
    elif 'Edam' in degree:
        df['What degree program are you currently enrolled in?'].replace(degree, 'Engineering', inplace=True)
    elif 'edam sharetime' in degree:
        df['What degree program are you currently enrolled in?'].replace(degree, 'Engineering', inplace=True)
    elif 'MET' in degree:
        df['What degree program are you currently enrolled in?'].replace(degree, 'Engineering', inplace=True)
    elif 'Mech tech' in degree:
        df['What degree program are you currently enrolled in?'].replace(degree, 'Engineering', inplace=True)
    elif 'Mec Tech' in degree:
        df['What degree program are you currently enrolled in?'].replace(degree, 'Engineering', inplace=True)    
    elif 'Information Security' in degree:
        df['What degree program are you currently enrolled in?'].replace(degree, 'ShareTime CSIP Program', inplace=True)
    elif 'Cybersecurity' in degree:
        df['What degree program are you currently enrolled in?'].replace(degree, 'ShareTime CSIP Program', inplace=True)
    elif 'Computer Science- Data Science' in degree:
        df['What degree program are you currently enrolled in?'].replace(degree, 'Computer Science', inplace=True)
    elif 'science and digital' in degree:
        df['What degree program are you currently enrolled in?'].replace(degree, 'Computer Science', inplace=True)    
    elif 'Data Science' in degree:
        df['What degree program are you currently enrolled in?'].replace(degree, 'Data Analytics Certificate of Achievement', inplace=True)
    elif 'high school' in degree:
        df['What degree program are you currently enrolled in?'].replace(degree, 'Challenger Program', inplace=True)
    elif 'Highschool' in degree:
        df['What degree program are you currently enrolled in?'].replace(degree, 'Challenger Program', inplace=True)
    elif 'MCVTS' in degree:
        df['What degree program are you currently enrolled in?'].replace(degree, 'Challenger Program', inplace=True)
    elif 'Morris Tech' in degree:
        df['What degree program are you currently enrolled in?'].replace(degree, 'Challenger Program', inplace=True)    
    else:
        print('not found')  # displays in each iteration whether a test string match was found
    

df['What degree program are you currently enrolled in?'].value_counts()  # displaying the updated condensed list

```
